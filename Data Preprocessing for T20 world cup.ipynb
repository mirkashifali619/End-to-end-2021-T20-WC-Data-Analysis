```python
import pandas as pd
import json
```

# Match Summary

```python
# opening the csv file
with open("t20_wc_match_results.json") as f:
    data = json.load(f)

df_summary = pd.DataFrame(data[0]['matchSummary'])
```

```python
df_summary.head()
```

```python
df_summary.rename({"scorecard":"match_id"} , inplace=True , axis=1)
```

```python
df_summary.head()
```

```python
df_summary.to_csv("fact_results_summary.csv" , index = False)
```

```python
match_ids_dict={}

for index, row in df_summary.iterrows():
    key1 = row['team1'] + " Vs " + row['team2']
    key2 = row['team2'] + " Vs " + row['team1']
    
    match_ids_dict[key1] = row["match_id"]
    match_ids_dict[key2] = row["match_id"]

match_ids_dict
```

# Batting Summary

```python
with open ("t20_wc_batting_summary.json") as f:
    data = json.load(f)
    
all_records=[]
for rec in data:
    all_records.extend(rec["battingSummary"])
    
df_batting = pd.DataFrame(all_records)
df_batting.head(10)
```

```python
df_batting["Out/not Out"] = df_batting.dismissal.apply(lambda x : "out" if len(x) > 0 else "Not Out")
df_batting.head(10)
```

```python
df_batting["batsmanName"] = df_batting["batsmanName"].apply(lambda x : x.replace("â€",""))
df_batting["batsmanName"] = df_batting["batsmanName"].apply(lambda x : x.replace("\xa0",""))
df_batting["dismissal"] = df_batting["dismissal"].apply(lambda x : x.replace("â€",""))
df_batting["dismissal"] = df_batting["dismissal"].apply(lambda x : x.replace("\xa0",""))
df_batting.head(10)
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
```

```python
df_batting
```

```python
df_batting.to_csv("fact_summary.csv", index  = False)
```

# Bowling Summary

```python
with open("t20_wc_bowling_summary.json") as f:
    data = json.load(f)
    
records_data=[]
for rec in data:
    records_data.extend(rec["bowlingSummary"])
    
df_bowling = pd.DataFrame(records_data)
```

```python
df_bowling.head(10)
```

```python
df_bowling["match_id"] = df_bowling["match"].map(match_ids_dict)
```

```python
df_bowling["bowlerName"] = df_bowling["bowlerName"].apply(lambda x : x.replace("â€",""))
df_bowling["bowlerName"] = df_bowling["bowlerName"].apply(lambda x : x.replace("\xa0",""))
```

```python
df_bowling
```

```python
df_bowling.to_csv("fact_bowling_summary.csv" , index = False)
```
